```
import networkx as nx
import matplotlib.pyplot as plt
```

```
H = nx.Graph()
```

```
H.add_node(1)
H.add_node(2)
H.add_node(3)
H.add_node(4)
H.add_node(5)
```

```
H.nodes()
```

```
H.add_edge(1,2)
H.add_edge(1,3)
H.add_edge(4,6)
H.add_edge(5,4)
H.add_edge(2,3)
```

```
H.edges()
```

```
nx.draw(H)  
```

```
nx.draw(H , with_labels = 1)  
```

```
H.add_edge(2,4)  # one more edge is added to connect the two different component of a Graph
```

```
nx.draw(H , with_labels = 1)  
```

#Complete graph  
 put all the possible edges between given vertices

```
# complete Grapg of 10 vertices where every node is connected with every vertices.
z = nx.complete_graph(10)  # all possible friendship between 10 people
```

```
z.nodes()
```

```
print(len(z.edges()))
z.edges()
```

```
z.order()  # Number. of nodes in a graph
```

```
z.size()  # size of the graph
```

```
nx.draw(z , with_labels = 1) # vertices with the labels
```

```
# complete graph with 100 vertices
G = nx.complete_graph(100)
nx.draw(G , with_labels = 1)
```

```
# complete graph with 100 vertices
G = nx.complete_graph(100)
plt.figure(figsize=(15,15))
nx.draw(G , with_labels = 1)
plt.show()
```

# GNp random graph    
the edges between every vertices are put/made based on probability       
In real life how two persons have friendship with each other (calculate it mathematically)      
**It is not mandatory here to put the edges between all vertices**   It means not comulsory here that each node should be connected with every nodes.

```
G = nx.gnp_random_graph(20 , 0.5)  # 20 vertices   # edges are randomly put (with 50% probability)
```

```
nx.draw(G , with_labels = 1)
```
