#2nd Example

```
import random
H = nx.Graph()
city_set1 = ['New York', 'Chicago', 'Boston',"Assam","Bihar","Chhattisgarh","Goa","Gujarat","Haryana"]
for each in city_set1:
    H.add_node(each)
print(H.size())    # size of the graph 
print(H.order())   # number of vertices in the graph 
H.nodes()
```

```
nx.draw(H , with_labels = 1 )
```

```
print(H.number_of_nodes())
print(H.number_of_edges())
costs = []
c = 100 
while(c<=2000):
    costs.append(c)
    c+=100
print(costs)
```

```
k = list(H.nodes())
k
```

```
while(H.number_of_edges()<10 ):
    c1 = random.choice(k)
    c2 = random.choice(k)
    if c1!=c2 and H.has_edge(c1 ,c2)==0:
        w = random.choice(costs)
        H.add_edge(c1,c2,weight = w)        
nx.draw(H , with_labels = 1)
        
```

```
pos = nx.spectral_layout(H)
nx.draw(H, pos,with_labels = 1)
```

```
pos = nx.circular_layout(H)
nx.draw(H, pos,with_labels = 1)
```

```
nx.is_connected(H)
```

```
for u in H.nodes():
    for v in H.nodes():
        print(u,v , nx.has_path(H,u,v))  # we csan't go anywhere from Newyork .
```

**Find the shortest path between two cities **

```
print(nx.dijkstra_path_length(H,'Goa','Bihar'))
nx.dijkstra_path(H,'Goa','Bihar')  
```

#Example 3

```
import networkx as nx 
import matplotlib.pyplot as plt 
import random 

G=nx.Graph() 
city_list=['a','b','c','d','e','f','g','h']

# this is a bit easier then adding each node in a loop 
G.add_nodes_from(city_list)
print(G.nodes())
print(type(G.nodes()))
G
```

```
# based on your old code:    
for _ in city_list: 
    ab=random.choice(city_list) 
    print(ab)
```

```
# generate n random edges
n=5
for _ in range(n):
    # random.sample(city_list, 2) gives a 2-tuple from city list
    # The '*'-operator unpacks the tuple to two input values for the .add_edge() method
    G.add_edge(*random.sample(city_list, 2))
print("Edges generated:", G.edges())
```

```
nx.draw(G , with_labels = 1)
```

```
pos = nx.spectral_layout(G)
nx.draw(G, pos,with_labels = 1)
```

```
pos = nx.circular_layout(G)
nx.draw(G, pos,with_labels = 1)
```

```
pos = nx.circular_layout(G)
nx.draw(G , with_labels = 1)
nx.draw_networkx_edge_labels(G,pos)
plt.show()
```

```
nx.is_connected(G)
```
