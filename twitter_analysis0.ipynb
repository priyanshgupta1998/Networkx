```
# Get them from http://dev.twitter.com
#put your own keys and tokens_id
OAUTH_TOKEN = " "
OAUTH_SECRET = " "
CONSUMER_KEY = " "
CONSUMER_SECRET = " "
```

```
print(dir(twitter))
```

```
# Log in
api = twitter.Api(consumer_key=CONSUMER_KEY,
                  consumer_secret=CONSUMER_SECRET,
                  access_token_key=OAUTH_TOKEN,
                  access_token_secret=OAUTH_SECRET)
```

```
# Load data
print("Loading users who follows",username,"...")
followers = []
followers.extend(api.GetFollowers(screen_name=username))
```

```
print(len(followers))
followers 
```

```
print("Loading users that",username,"follows...")
friends = []
friends.extend(api.GetFriends(screen_name=username))
```

```
print(len(friends))
friends[:10]
```

```
# Create graph
print("Adding followers relationships...")
for user in followers:
	graph.add_edge(user.name,username)
```

```
print(len(graph))
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
nx.draw_networkx(graph)
plt.show()   # directed graph 
```

```
print("Adding following relationships that  ", username , " is following  others.......")
for user in friends:
	graph.add_edge(username,user.name)
```

```
print(len(graph))
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
nx.draw_networkx(graph)
plt.show()   # directed graph   
```
