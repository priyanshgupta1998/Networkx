```
import networkx as nx
import matplotlib.pyplot as plt
```

```
G =nx.read_edgelist('/home/facebook_combined.txt')
G
```

```
print(nx.info(G))
print('\n')
print(nx.is_directed(G))
```

```
P = nx.read_pajek('/home/football.net')
P
```

```
print(nx.info(P))
print('\n')
print(nx.is_directed(P))
```

```
K = nx.read_graphml('/home/wikipedia.graphml')
K
```

```
print(nx.info(K))
print('\n')
print(nx.is_directed(K))
```

```
L = nx.read_gml('/home/karate.gml')
L
```

```
print(nx.info(L))
print('\n')
print(nx.is_directed(L))
```

```
nx.draw(L , with_labels= 1 )
```

```
nx.draw_circular(L , with_labels= 1)
```

```
nx.degree(L)
```

```
all_degrees = list(dict(nx.degree(L)).values())
all_degrees
```

```
print(set(dict(nx.degree(L)).values()))  # not repeated values in order.
print(list(set(dict(nx.degree(L)).values())))
```

```
unique_degrees = list(set(dict(nx.degree(L)).values()))
count_of_degrees = []
for i in unique_degrees:
    x = all_degrees.count(i)
    count_of_degrees.append(x)
plt.plot(unique_degrees ,count_of_degrees,'go-' )
   
```

```
plt.loglog(unique_degrees ,count_of_degrees,'rs-' ) 
```

```
print('Density of karate network ' , nx.density(L))
```

# Density of the Graph   
 D = number of edges present in the graph / total possible edges    
 

```
G = nx.complete_graph(100)
nx.density(G)
```

```
H = nx.Graph()
H.add_nodes_from([1,2,3,4])
nx.density(H)
```
